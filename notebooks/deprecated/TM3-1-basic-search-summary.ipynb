This is a Jupyter notebook for quickly summarizing the status of a HOTOSM Tasking Manager campaing using the TM API.
It uses the same search functionality as the keyword search o tasks.atotosm.org/contribute

**Usage**:

1. Set your search keyword to the variable `textSearch`
2. Execute all cells.
3. The results are sorted by project ID by default so the order stays consistent over time

```python
# set the text search keyword
textSearch = 'COVID'
sort_by = 'projectId'

#################
import requests
import json
import pandas as pd

def getAllPages(textSearch):
    page = 1
    while True:
        r = requests.get(f'https://tasks.hotosm.org/api/v1/project/search?mapperLevel=ALL&textSearch={textSearch}&page={page}', 
            headers = {
                'Accept-Language': '*'
            }
        )
        j = json.loads(r.content)
        df = pd.DataFrame(json.loads(r.content)['results'])
        yield df
        if j['pagination']['hasNext'] is True:
            page = j['pagination']['nextNum']
        else:
            break

res = pd.concat(getAllPages(textSearch), ignore_index=True);

pd.set_option('display.max_rows', 500)
display(res[['projectId','name','percentMapped', 'percentValidated','priority','status']].sort_values(sort_by))
```

```python
# prints the above pretty-printed data in plain CSV format 
from io import StringIO
csv = StringIO()
del res['shortDescription']
res.to_csv(csv, index=False)
print(csv.getvalue());
```
