This is a Jupyter notebook for reviewing users of mapped tasks of a specific project in the Tasking Manager.

**Usage**:

1. Set the `projectID` of the project you are interested in.
2. Click the play button.


```python
# set the projectiD
projectId = '5660'
sort_by = 'dateRegistered'
ascending = False
instance = 'tasking-manager-staging-api.hotosm.org'
# instance = 'tasking-manager-tm4-api.hotosm.org'


import requests, json, pandas
def getData(projectId):
    query = f'https://{instance}/api/v2/projects/{projectId}/tasks/queries/mapped/'
    result = requests.get(query, headers = {
        'Accept-Language': '*',
        'Content-Type': 'application/json'
    })
                         
    j = json.loads(result.content.decode('utf8'))
    df = pandas.DataFrame(j['mappedTasks'])
    return df

result = getData(projectId);
display(result[['username','mappedTaskCount','mappingLevel', 'dateRegistered','lastValidationDate']].sort_values(sort_by, ascending = ascending))
```
